```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.ticker as ticker
```

```python
df=pd.read_excel('150projectDATA.xlsx')

# splitting the data into training (60%) and testing (40%)
train_data = df.iloc[:32,1]
test_data = df.iloc[33:,1]
```

```python
# performing a grid search
errors = []
models = []
parameters = []
for i in range(5):
    for j in range(5):
        for k in range(5):
            model = ARIMA(train_data,order=(i+1,j+1,k+1))
            model_fit = model.fit()
            # testing the model
            forecast = model_fit.forecast(len(test_data))
            mse = np.mean((test_data - forecast)**2)
            errors.append(mse)
            models.append(model_fit)
            parameters.append((i+1,j+1,k+1))
        
print(f'minimum test mse: {min(errors)}')
```

```python
x = errors.index(min(errors))
print(errors[x], models[x], parameters[x])
```

```python
# plotting the best tested model
model = models[x]
forecast = model.forecast(len(test_data))

plt.plot(df.iloc[33:,0],test_data)
plt.plot(df.iloc[33:,0],forecast)
# plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: '{:.0f}'.format(forecast)))
plt.legend(['ground truth','tested forecast'])
plt.xlabel('time')
plt.ylabel('Return')
plt.title('Rate of Return ARIMA Prediction')
plt.show()

# plotting the training progress of our model
train_preds = model.predict()
plt.plot(df.iloc[:32,0],df.iloc[:32,1])
plt.plot(df.iloc[:32,0],train_preds)
plt.legend(['ground truth','trained forecast'])
plt.xlabel('time')
plt.ylabel('Return')
plt.title('Rate of Return ARIMA Prediction')
plt.show()
```
